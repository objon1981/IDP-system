```python
import subprocess

# Load the requirements
with open("requirements.txt", "r") as f:
    lines = f.readlines()

new_lines = []

print("🛠️ Processing requirements...")

for line in lines:
    req = line.strip()

    # Skip empty lines and already-commented lines
    if not req or req.startswith("#"):
        new_lines.append(line)
        continue

    print(f"🔍 Trying: {req}")
    try:
        subprocess.check_output(["pip", "install", req], stderr=subprocess.STDOUT)
        new_lines.append(line)
    except subprocess.CalledProcessError as e:
        print(f"❌ Could not install: {req} — Commenting it out.")
        new_lines.append(f"# {req}  # Could not install\n")

# Save updated file
with open("requirements.txt", "w") as f:
    f.writelines(new_lines)

print("\n✅ Done. Final requirements:")
!cat requirements.txt
```

```python
import pkg_resources
import re

# Step 1: Read and extract package names from requirements.txt
def get_requirements(requirements_path="requirements.txt"):
    with open(requirements_path, "r") as f:
        lines = f.readlines()

    reqs = []
    for line in lines:
        line = line.strip()
        if not line or line.startswith("#"):
            continue
        pkg = re.split("[>=<]", line)[0].lower().replace("_", "-")
        reqs.append(pkg)
    return set(reqs)

# Step 2: Get installed packages
installed_packages = {pkg.key for pkg in pkg_resources.working_set}

# Step 3: Compare
required_packages = get_requirements()
not_installed = required_packages - installed_packages

# Display results
print("🚫 Packages from requirements.txt that are still NOT installed:\n")
for pkg in sorted(not_installed):
    print(f"- {pkg}")
```
